```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch, Rectangle
import matplotlib.patheffects as path_effects

url = "https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/storms.csv"
df_storms = pd.read_csv(url)

years = (2011, 2012)
df1 = df_storms[df_storms["year"].isin(years)]
df1[(df1["year"]==2011)] ["n"].sum()
df1[(df1["year"]==2012)] ["n"].sum()

colors = ["#003f5c", "#7a5195", "#ef5675", "#ffa600"]   

fig,ax = plt.subplots()
plt.subplots_adjust(top=0.8, left=0.22, right=0.77)

ax.set_xlim([0,30])
ax.set_ylim([0,60])

verts = [(0, 57.4), (15, 57.4), (15, 15.8), (30, 15.8), (30, 3.6), (15, 3.6),
         (15, 46.8), (0, 46.8), (0, 57.4)]
codes = [Path.MOVETO, Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.LINETO,
         Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.CLOSEPOLY]
p = Path(verts,codes)
ax.add_patch(PathPatch(p, alpha = 0.9, color=colors[0]))
ax.text(x=-5.7, y=53.1, s="hurricane", color=colors[0])
ax.text(x=-2.5, y=50.1, s=106, color=colors[0])
ax.text(x=30, y=9.7, s=122, color=colors[0])

verts = [(0, 44.8), (15, 44.8), (15, 24.8), (30, 24.8), (30, 17.8), (15, 17.8),
         (15, 40.1), (0, 40.1), (0, 44.8)]
codes = [Path.MOVETO, Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.LINETO,
         Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.CLOSEPOLY]
p = Path(verts,codes)
ax.add_patch(PathPatch(p, alpha = 0.9, color =colors[1]))
ax.text(x=-11.5, y=42.45, s="tropical depression", color=colors[1])
ax.text(x=-2, y=39.45, s=47, color=colors[1])
ax.text(x=30, y=21.3, s=70, color=colors[1])

verts = [(0, 38.1), (15, 38.1), (15, 55.4), (30, 55.4), (30, 26.8), (15, 26.8),
         (15, 12.5), (0, 12.5), (0, 38.1)]
codes = [Path.MOVETO, Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.LINETO,
         Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.CLOSEPOLY]
p = Path(verts,codes)
ax.add_patch(PathPatch(p, alpha = 0.9, color =colors[2]))
ax.text(x=-8.3, y=25.3, s="tropical storm", color=colors[2])
ax.text(x=-2.5, y=22.3, s=256, color=colors[2])
ax.text(x=30, y=41.1, s=286, color=colors[2])

verts = [(0, 10.5), (15, 10.5), (15, 1.6), (30, 1.6), (30, 0), (15, 0),
         (15, 9.8), (0, 9.8), (0, 10.5)]
codes = [Path.MOVETO, Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.LINETO,
         Path.CURVE4, Path.CURVE4, Path.CURVE4, Path.CLOSEPOLY]
p = Path(verts,codes)
ax.add_patch(PathPatch(p, alpha = 0.9, color =colors[3]))
ax.text(x=-8, y=10.15, s="tropical wave", color=colors[3])
ax.text(x=-1.5, y=7.15, s=7, color=colors[3])
ax.text(x=30, y=0.2, s=16, color=colors[3])

ax.axvline(x=0, lw=4, color="darkgrey")
ax.axvline(x=30, lw=4, color="darkgrey")

location=[0, 30]
labels=[2011, 2012]
ax.set_xticks(location, labels=labels)

year = df1['year'].unique()

for years, xticks in zip(year, location):
    ax.text(s=years, x=xticks, y=62, color="white", weight="bold", clip_on=False,
            ha="center", bbox=dict(boxstyle='round,pad=0.8', facecolor="black"))

t = fig.text(0.5, 0.92, "Tropical storms", size=30, weight="bold", ha="center")
t.set_path_effects([
    path_effects.PathPatchEffect(
        offset=(2, -2), hatch='xxxx', facecolor='gray'),
    path_effects.PathPatchEffect(
        edgecolor='white', linewidth=1.1, facecolor='black')])

ax.set_axis_off()

fig.text(0.5, 0.02, "Data: Matplotlib Journey - Storms | Design: Bernadett Piros",
         size=6, ha="center")

plt.show()

#plt.savefig("sankey_storms", dpi=300, bbox_inches='tight', pad_inches=0.2)
```
