```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pypalettes import load_cmap

url = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/wine/wine.csv"
df = pd.read_csv(url)

df.columns = df.columns.str.replace("_", " ")
df = df.groupby("quality").mean().reset_index()
df = df.set_index("quality")

plt.style.use("Solarize_Light2")
cmap = load_cmap("Ronda")

x = np.arange(len(df.columns))
y = np.arange(len(df.index))
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots()
ax.contourf(X, Y, df.values, cmap=cmap)

location_y = list(range(0,7))
labels_y = df.index.tolist()
ax.set_yticks(location_y, labels=labels_y)
ax.tick_params(axis="y", labelsize=8)
ax = plt.gca()
for tick in ax.get_yticklabels():
    tick.set_fontweight('bold')

ax.set_xlabel("Characteristics", size=8, weight="bold")
ax.set_xticks([])

ax.tick_params(length=0)
ax.spines[["top", "bottom", "right", "left"]].set_visible(False)

fig.text(0.5, 0.95, 'What distinguishes a "good" wine from a "bad" one?\nYour personal taste!',
         size=14, ha="center", va="center", fontweight="bold", color="#49295FFF")

fig.text(0.01, 0.8, "Quality", color="#6d8087", size=8, fontweight="bold")

fig.text(0.01, 0.03, df.columns.tolist(), size=6, color="#6d8087")

fig.text(0.5, -0.02, "Data: Matplotlib Journey - Wine | Design: Bernadett Piros",
         size=6, ha="center", color="#49295FFF")
  
plt.show()

#plt.savefig("contour_wine")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pypalettes import load_cmap
from PIL import Image

url = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/wine/wine.csv"
df = pd.read_csv(url)

df.columns = df.columns.str.replace("_", " ")
df = df.groupby("quality").mean().reset_index()
df = df.set_index("quality")

plt.style.use("Solarize_Light2")
cmap = load_cmap("Ronda")

x = np.arange(len(df.columns))
y = np.arange(len(df.index))
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots()
ax.contourf(X, Y, df.values, cmap=cmap)

location_y = list(range(0,7))
labels_y = df.index.tolist()
ax.set_yticks(location_y, labels=labels_y)
ax.tick_params(axis="y", labelsize=8)
ax = plt.gca()
for tick in ax.get_yticklabels():
    tick.set_fontweight('bold')

ax.set_xlabel("Characteristics", size=8, weight="bold")
ax.set_xticks([])

ax.tick_params(length=0)
ax.spines[["top", "bottom", "right", "left"]].set_visible(False)

fig.text(0.5, 0.95, 'What distinguishes a "good" wine from a "bad" one?\nYour personal taste!',
         size=14, ha="center", va="center", fontweight="bold", color="#49295FFF")

fig.text(0.01, 0.8, "Quality", color="#6d8087", size=8, fontweight="bold")

fig.text(0.01, 0.03, df.columns.tolist(), size=6, color="#6d8087")

fig.text(0.4, -0.02, "Data: Matplotlib Journey - Wine | Design: Bernadett Piros |",
         size=6, ha="center")

def add_logo(path_to_logo, text, image_bottom_right_x, image_bottom_right_y, image_width):
    logo = Image.open(path_to_logo) 
    image_array = np.array(logo) 
    image_height = image_width * image_array.shape[0] / image_array.shape[1]  
    
    ax_image = plt.axes([image_bottom_right_x, 
                         image_bottom_right_y, 
                         image_width, 
                         image_height]) 
    ax_image.imshow(image_array) 
    ax_image.axis('off')     
    
    fig.text(0.62, 
             image_bottom_right_y+0.01, 
             text,
             fontsize=6
             )

add_logo("Logo/github-mark.png", "DevJupyHub", 0.59, -0.03, 0.03)
  
plt.show()

#plt.savefig("gcontour_wine")
```
